```python
# Merge WDI and WGI data into a single country-year panel
df = wdi.merge(wgi, on=['country', 'year'], how='left')
df['country'] = df['country'].astype(str)

# Tag Francophone countries from the list
fr_countries = set(francop['country'].str.strip())
df['is_francophone_country'] = df['country'].isin(fr_countries).astype(int)

# Sort by country and year to prepare for time-series calculations
df = df.sort_values(['country', 'year'])

# Compute previous year population and population growth rate
df['pop_prev'] = df.groupby('country')['pop_total'].shift(1)
df['pop_growth_pct'] = (df['pop_total'] / df['pop_prev'] - 1.0) * 100.0

# Compute a GeoRisk index (mean of all WGI governance scores)
wgi_cols = ['pol_stability','voice_account','gov_effectiveness','reg_quality','rule_of_law','control_corruption']
df['georisk_index'] = df[wgi_cols].mean(axis=1)

# Anglosphere flag for geopolitical contrast
anglo_countries = {'United States','United Kingdom','Canada','Australia','New Zealand'}
df['is_anglo'] = df['country'].isin(anglo_countries).astype(int)

df.head(10)
```

In this project, the dataset encodes historical and scenario-based geopolitical indicators into a structured numeric format (CSV). Each row corresponds to a time point (100 CE, 1750, 1950, 2020, and 2070), and each column represents a dimension of international power that influences language dominance. For example, state reach measures how much territory and population are unified under one political system, while military projection captures the ability to project power beyond borders. Trade centrality reflects whether the hegemon is positioned at the center of major trade routes, and monetary influence measures how widely its currency is accepted for trade and reserves. Market depth goes a step further: it refers to the size and liquidity of a hegemon’s financial markets and their ability to set global prices and provide capital — for instance, the New York Stock Exchange today or Rome’s role in Mediterranean grain financing. Ideology hegemony measures how much the leading state’s worldview (e.g., Roman law, Enlightenment values, American liberal capitalism) dominates international norms, while language prestige (separately tracked for Latin, French, and English) captures how much each language is used in diplomacy, finance, and culture. Emerging market weight measures how economically significant rising regions are (e.g., Egypt in Roman times, Africa in the future), and emerging market language adoption records how much these regions adopt the hegemon’s language for administration, trade, and education. Finally, emerging market financial linkages assess whether these regions are plugged into the hegemon’s financial hubs or operate in isolation.

```python
# Build a 2022 panel for Francophone countries to allocate global French figures
anchor_year = 2022
oif_total = int(oif.loc[oif['metric'] == 'french_speakers_global_total', 'value'])
oif_daily = int(oif.loc[oif['metric'] == 'daily_french_users_global', 'value'])

panel_2022 = df[(df['year'] == anchor_year) & (df['is_francophone_country'] == 1)].copy()
panel_2022['pop_share_fr'] = panel_2022['pop_total'] / panel_2022['pop_total'].sum()

# Compute a quality factor for each country (normalized governance + education metrics)
quality_vars = ['voice_account','gov_effectiveness','reg_quality','rule_of_law','control_corruption','edu_spend_gdp_pct']
q = panel_2022[quality_vars].copy()
q = (q - q.min()) / (q.max() - q.min())  # min-max normalize each variable
panel_2022['quality'] = q.mean(axis=1).fillna(q.mean().mean())

# Allocation weight = population share * blended quality (60% pop share + 40% quality)
panel_2022['alloc_weight'] = panel_2022['pop_share_fr'] * (0.6 + 0.4 * panel_2022['quality'])
panel_2022['alloc_weight'] /= panel_2022['alloc_weight'].sum()  # normalize weights to 1

# Allocate global totals to each country based on alloc_weight
panel_2022['allocated_french_speakers'] = (oif_total * panel_2022['alloc_weight']).round().astype('Int64')
panel_2022['allocated_daily_users']    = (oif_daily * panel_2022['alloc_weight']).round().astype('Int64')

# Use panel_2022 as the base for modeling dataset (only Francophone countries in 2022)
df_modeling = panel_2022.copy()

# Construct a per-capita influence proxy: daily French users as fraction of population
df_modeling['french_influence_pc'] = df_modeling['allocated_daily_users'] / df_modeling['pop_total']
df_modeling['french_influence_pc'] = df_modeling['french_influence_pc'].clip(upper=1.0)  # cap at 1.0 (100%)

df_modeling[['country','year','is_francophone_country','french_influence_pc']].head()

df.head()
```

## Target Construction Using OIF Anchors
OIF provides **global** counts (321M speakers in 2022; 212M daily users; 54.7% of daily users in Africa). We construct a *country‑year proxy* target by allocating global anchors proportionally based on population shares within **Francophone countries** and scaling by governance/education to reflect usable French influence.

Superposition of Outcomes: A qubit in superposition can represent multiple possible states at once. Why Quantum rather than a Neural Network?
It can offer advantages in specific situations because it processes information in fundamentally different ways. A neural network learns patterns using classical arithmetic operations, while a quantum circuit represents data using quantum states that can exist in superposition and become entangled. This allows quantum models to explore extremely high-dimensional feature spaces using only a small number of qubits, potentially capturing complex correlations that classical models would need many more parameters to represent. In problems where patterns depend on subtle interference effects or highly nonlinear relationships, a quantum circuit can in principle encode and transform data in ways that classical networks cannot reproduce efficiently. However, quantum models also tend to be harder to train, more sensitive to noise, and not yet consistently superior in real-world tasks. So the advantage isn’t that they “predict better” universally, but that they can sometimes learn richer transformations from fewer parameters by leveraging quantum superposition and entanglement—properties no classical neural network can emulate directly.
